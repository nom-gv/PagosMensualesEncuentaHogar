```python
#Librerias
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import r2_score
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
#Importar datos
url = "https://raw.githubusercontent.com/nom-gv/DatasetPersona/main/Base2019.csv"
df1 = pd.read_csv(url)
df1
```

```python
#Importar datos
url = "https://raw.githubusercontent.com/nom-gv/DatasetPersona/main/Base2018.csv"
df2 = pd.read_csv(url)
df2
```

```python
#Importar datos
url = "https://raw.githubusercontent.com/nom-gv/DatasetPersona/main/Base2017.csv"
df3 = pd.read_csv(url)
df3
```

```python
#Seleccionar variables
x1 = "nro_residentes"
x2 = "tipo_vivienda"
y = "pago_mensual"
```

```python
#Crear resultados
variables_x1 = [x1, x2]
variable_y1 = y
modelo1 = LinearRegression() # generamos la regresión lineal
modelo1.fit(df1[variables_x1], df1[variable_y1]) # entrenamos el modelo
print ('Coeficientes: ', modelo1.coef_) #obtenemos los coeficientes de las variables independientes
print ('Intercepción: ', modelo1.intercept_) #obtenemos la intercepción
# Imprimir la ecuación del plano
print('Ecuación del plano: y = {} * nro residentes + {} * tipo vivienda + {}'.format(round(modelo1.coef_[0], 3), round(modelo1.coef_[1], 3), round(modelo1.intercept_, 3)))
print ('Coeficiente de determinación: ', round(r2_score(df1[variable_y1], modelo1.predict(df1[variables_x1])), 3))
```

```python
# Cargar datos
data = {
    "anio":[2014,2015,2016,2017,2018,2019],
    "pago_mensual":[732,775,794,812,851,881]
}
df4 = pd.DataFrame(data)
df4
```

```python
# Seleccionar variables
variable_x = "anio"
variable_y = "pago_mensual"
```

```python
# Generar análisis
modelo = LinearRegression() # generamos la regresión lineal
modelo.fit(df4[[variable_x]], df4[variable_y]) # entrenamos el modelo
print ('Ecuación de la recta: y = ', round(modelo.coef_[0],3),'x + ', round(modelo.intercept_,3))#obtenemos la ecuación de la recta
print ('Coeficiente de correlación: ', round(np.corrcoef(df4[variable_x], df4[variable_y])[0,1], 3))#obtenemos el coeficiente de correlación
print ('Coeficiente de determinación: ', round(r2_score(df4[variable_y], modelo.predict(df4[[variable_x]])), 3))#obtenemos el coeficiente de determinación
# Gráfica con intervalo de confianza
sns.regplot(x=df4[variable_x], y=df4[variable_y], ci=95, line_kws={"color": "red"}, scatter_kws={"color": "black"})
plt.title('Regresión lineal simple con intervalo de confianza al 95%')
plt.show()
```

```python
# Generar PREDICCIÓN
dato_predictor = 2023

anio = pd.DataFrame([dato_predictor], columns=[variable_x])  # Convertir a DataFrame
pago = modelo.predict(anio)
print('La predicción de pago mensual para', anio.iloc[0][0], 'anios es', round(pago[0], 3))

dato_predictor1 = 2022

anio1 = pd.DataFrame([dato_predictor1], columns=[variable_x])  # Convertir a DataFrame
pago1 = modelo.predict(anio1)
```

```python
pago_mes = round(pago1[0], 3)
pago_mes1 = round(pago[0], 3) - pago_mes
porcentaje = (100*pago_mes1)/(pago_mes)
print(porcentaje)
```
